# Problemas de Estadística Descriptiva

## Ejercicio 1 – ¿Qué tan largos son los títulos de las películas?

En este ejercicio vas a aplicar técnicas de estadística descriptiva sobre un conjunto de datos real: una base con 1000 películas populares del sitio IMDb.

A partir del título de cada película, vas a:
- Calcular la longitud del título (en caracteres).
- Obtener métricas estadísticas clave: media, mediana, moda, rango, varianza, desviación estándar, asimetría y curtosis.
- Identificar los títulos más largos y más cortos.
- Visualizar la distribución de estas longitudes con un histograma y una curva de densidad.


```python
# - Importe De Librerías a Emplear:

import pandas as pd
import random
import math
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from scipy.stats import skew, kurtosis, mode
import statistics as stats

# - Carga Del DataSet Desde IMDB:

url = "../assets/imdb_1000.csv"
df = pd.read_csv(url)
df
```

```python
# - Creación de una nueva columna, titulada: "title_length", donde esté la cantidad de carácteres en cada título:

df["title_length"] = df['title'].str.len()
df
```

```python
# - Cálculo de la media:

mean = stats.mean(df['title_length'])
print(f"El cálculo de la media da como resultado: {round(mean, 2)}")
print()

# - Cálculo de la mediana:

median = stats.median(df['title_length'])
print(f"El cálculo de la mediana da como resultado: {median}")

# - Cálculo de la moda:

print()
mode = stats.mode(df['title_length'])
print(f"El cálculo de la moda da como resultado: {mode}")
```

```python
# - Cálculo del rango:

range = max(df['title_length']) - min(df['title_length'])
print(f"El cálculo del rango ha dado como resultado: {range}")
print()

# - Cálculo de la varianza:

variance = stats.variance(df['title_length'])
print(f"El cálculo de la varianza ha dado como resultado: {round(variance, 2)}")

# - Cálculo de la desviación estándar:

print()
std = stats.stdev(df['title_length'])
print(f"El cálculo de la desviación estándar ha dado como resultado: {round(std, 2)}")
```

```python
# - Cálculo de la asimetría:

skewness = skew(df['title_length'])
print(f"El cálculo de la asimetría ha dado como resultado: {round(skewness, 2)}")
print()

# - Cálculo de la curtosis:

kurt = kurtosis(df['title_length'])
print(f"El cálculo de la curtosis ha dado como resultado: {round(kurt, 2)}")
```

```python
# - Impresión del título más corto:

mint = df['title_length'].idxmin()
mintt = df.loc[mint, 'title']
print(f"El título más corto es: {mintt}")
print()

# - Impresión del título más largo:

maxt = df['title_length'].idxmax()
maxtt = df.loc[maxt, 'title']
print(f"El título más largo es: {maxtt}")
```

```python
# - Graficación de todo lo obtenido hasta ahora:

fig, axes = plt.subplots(1, 3, figsize=(18, 6))

# - Histograma con Asimetría & Curtosis:

sns.histplot(df['title_length'], kde=True, ax=axes[0], color='red')
axes[0].set_xlabel("Número De Carácteres En El Título")
axes[0].set_ylabel("Conteo")
axes[0].set_title("Histograma: Asimetría & Curtosis")

# - Boxplot con el rango:

sns.boxplot(x=df['title_length'], ax=axes[1], color='lightgreen')
axes[1].set_xlabel("Número De Carácteres En El Título")
axes[1].set_ylabel("Conteo")
axes[1].set_title("Boxplot: Rango")

# - Histograma con Media, mediana & moda:

sns.histplot(df['title_length'], ax=axes[2], color='royalblue')
axes[2].axvline(df['title_length'].mean(), color='blue', label='Media')
axes[2].axvline(df['title_length'].median(), color='aqua', label='Mediana')
axes[2].axvline(df['title_length'].mode()[0], color='navy', label='Moda')
axes[2].set_xlabel("Número De Carácteres En El Título")
axes[2].set_ylabel("Conteo")
axes[2].set_title("Histograma: Media, Mediana & Moda")
axes[2].legend()

# - Visualización:

plt.tight_layout()
plt.show()
```

## Ejercicio 2 – Cálculo manual de desviación estándar

En este ejercicio vas a calcular manualmente la desviación estándar, seleccionando cinco títulos reales del dataset de películas que cargaste antes.


```python
# - Recopilación de cinco valores de la columna "title_length":

random5 = random.sample(list(df['title_length']), 5)
print("Valores seleccionados:", random5)
```

```python
# - Media de los cinco valores:

media = sum(random5)/5
print(media)
```

```python
# - Cálculo de las diferencias cuadráticas respecto a la media:

sum_diferencias_cuadrado = sum((x - media) ** 2 for x in random5)
print(sum_diferencias_cuadrado)
```

```python
# - Cálculo de la desviación estándar:

varianza = sum_diferencias_cuadrado / (len(random5) - 1)
sigma = math.sqrt(varianza)
print("Desviación Estándar:", round(sigma, 4))
```
